```python
def read_input(filename):
    sections = []
    with open(filename, 'r') as f:
        for line in f.readlines():
            sections.append([[int(c) for c in s.split('-')] for s in line.strip().split(',')])
    return sections

def range_covers(ranges):
    r1, r2 = ranges
    return (r1[0] <= r2[0] and r1[1] >= r2[1]) or (r2[0] <= r1[0] and r2[1] >= r1[1])

def range_overlaps(ranges):
    r1, r2 = ranges
    return (r1[1] >= r2[0] and r1[0] <= r2[1]) or (r2[1] >= r1[0] and r2[0] <= r1[1])

def runit(filename, part=1):
    ranges = read_input(filename)
    func = range_covers if part == 1 else range_overlaps
    return sum(func(r) for r in ranges)
```

```python
ranges = read_input("04_input.txt")
```

```python
[range_covers(range) for range in ranges]
```

```python
runit('04_input.txt')
```

```python
runit('04_input.txt', part=2)
```
