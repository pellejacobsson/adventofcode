```python
from itertools import cycle
import numpy as np
```

```python
class Piece1:
    def __init__(self, chamber):
        y = max(chamber, key=lambda k: k[1])[1] + 4
        self.x = 2
        self.y = y
    
    def move(self, chamber, jet):
        dir = next(jet)
        moved = True
        if dir == '>':
            if (self.x + 4, self.y) not in chamber and self.x + 3 < 6:
                self.x += 1
        elif dir == '<':
            if (self.x - 1, self.y) not in chamber and self.x > 0:
                self.x -= 1
        else:
            print('Wrong direction 1')
        if not any((x, self.y - 1) in chamber for x in range(self.x, self.x + 4)):
            self.y -= 1
        else:
            moved = False
            for x in range(self.x, self.x + 4):
                chamber[x, self.y] = True
        return moved

class Piece2:
    def __init__(self, chamber):
        y = max(chamber, key=lambda k: k[1])[1] + 4
        self.x = 3
        self.y = y

    def move(self, chamber, jet):
        dir = next(jet)
        moved = True
        if dir == '>':
            if not ((self.x + 1, self.y) in chamber or (self.x + 2, self.y + 1) in chamber or (self.x + 1, self.y + 2) in chamber) and self.x + 1 < 6:
                self.x += 1
        elif dir == '<':
            if not ((self.x - 1, self.y) in chamber or (self.x - 2, self.y + 1) in chamber or (self.x - 1, self.y + 2) in chamber) and self.x - 1 > 0:
                self.x -= 1
        else:
            print('Wrong direction 2')
        if not((self.x - 1, self.y) in chamber or (self.x, self.y - 1) in chamber or (self.x + 1, self.y) in chamber):
            self.y -= 1
        else:
            moved = False
            for x in range(self.x - 1, self.x + 2):
                chamber[x, self.y + 1] = True
            chamber[self.x, self.y] = True
            chamber[self.x, self.y + 2] = True
        return moved

class Piece3:
    def __init__(self, chamber):
        y = max(chamber, key=lambda k: k[1])[1] + 4
        self.x = 2
        self.y = y
    
    def move(self, chamber, jet):
        dir = next(jet)
        moved = True
        if dir == '>':
            if not ((self.x + 3, self.y) in chamber or (self.x + 3, self.y + 1) in chamber or (self.x, self.y + 2) in chamber) and self.x + 2 < 6:
                self.x += 1
        elif dir == '<':
            if not((self.x - 1, self.y) in chamber or (self.x + 1, self.y + 1) in chamber or (self.x + 1, self.y + 2) in chamber) and self.x > 0:
                self.x -= 1
        else:
            print('Wrong direction 3')
        if not((self.x, self.y - 1) in chamber or (self.x + 1, self.y - 1) in chamber or (self.x + 2, self.y - 1) in chamber):
            self.y -= 1
        else:
            moved = False
            for x in range(self.x, self.x + 3):
                chamber[x, self.y] = True
            chamber[self.x + 2, self.y + 1] = True
            chamber[self.x + 2, self.y + 2] = True
        return moved

class Piece4:
    def __init__(self, chamber):
        y = max(chamber, key=lambda k: k[1])[1] + 4
        self.x = 2
        self.y = y

    def move(self, chamber, jet):
        dir = next(jet)
        moved = True
        if dir == '>':
            if not any((self.x + 1, y) in chamber for y in range(self.y, self.y + 4)) and self.x < 6:
                self.x += 1
        elif dir == '<':
            if not any((self.x - 1, y) in chamber for y in range(self.y, self.y + 4)) and self.x > 0:
                self.x -= 1
        else:
            print('Wrong direction 4')
        if (self.x, self.y - 1) not in chamber:
            self.y -= 1
        else:
            moved = False
            for y in range(self.y, self.y + 4):
                chamber[self.x, y] = True
        return moved

class Piece5:
    def __init__(self, chamber):
        y = max(chamber, key=lambda k: k[1])[1] + 4
        self.x = 2
        self.y = y

    def move(self, chamber, jet):
        dir = next(jet)
        moved = True
        if dir == '>':
            if not ((self.x + 2, self.y) in chamber or (self.x + 2, self.y + 1) in chamber) and self.x + 1 < 6:
                self.x += 1
        elif dir == '<':
            if not((self.x - 1, self.y) in chamber or (self.x - 1, self.y + 1) in chamber) and self.x > 0:
                self.x -= 1
        else:
            print('Wrong direction 5')
        if not((self.x, self.y - 1) in chamber or (self.x + 1, self.y - 1) in chamber):
            self.y -= 1
        else:
            moved = False
            chamber[self.x, self.y] = True
            chamber[self.x + 1, self.y] = True
            chamber[self.x, self.y + 1] = True
            chamber[self.x + 1, self.y + 1] = True
        return moved
            

def read_input(filename):
    with open(filename, 'r') as f:
        jet = [s for s in f.read().strip()]
    return cycle(jet)

def print_chamber(chamber):
    ymax = max(chamber, key=lambda k: k[1])[1]
    for y in range(ymax + 7, -1, -1):
        line = ''.join(['#' if (x, y) in chamber else '.' for x in range(7)])
        print(line)

def runit(filename):
    jet = read_input('17_input.txt')
    chamber = {(x, -1): True for x in range(7)}
    for round in range(1, 2022 + 1):
        match round % 5:
            case 1:
                p = Piece1(chamber)
            case 2:
                p = Piece2(chamber)
            case 3:
                p = Piece3(chamber)
            case 4:
                p = Piece4(chamber)
            case 0:
                p = Piece5(chamber)
        moved = True
        while moved:
            moved = p.move(chamber, jet)
    return max(chamber, key=lambda k: k[1])[1] + 1

def calc_height(round, hh, dh):
    ix = round - 1
    h0 = len(hh)
    n_dh = len(dh)
    s_dh = sum(dh)
    if round <= h0:
        return hh[ix]
    n = (ix - h0) // n_dh
    m = (ix - h0) % n_dh
    h = hh[-1] + s_dh * n + sum(dh[:m+1])
    return h

def get_pattern(h):
    dh = np.diff(h)
    break_all = False
    for i in range(4000):
        if break_all:
            break
        for j in range(i+1000, 4000):
            if i != j and all(dh[i:i+1000] == dh[j:j+1000]):
                ix, jx = i, j
                break_all = True
                break
    return ix, jx - ix

def runit2(filename):
    jet = read_input('17_input.txt')
    chamber = {(x, -1): True for x in range(7)}
    h = []
    for round in range(1, 5000 + 1):
        match round % 5:
            case 1:
                p = Piece1(chamber)
            case 2:
                p = Piece2(chamber)
            case 3:
                p = Piece3(chamber)
            case 4:
                p = Piece4(chamber)
            case 0:
                p = Piece5(chamber)
        moved = True
        while moved:
            moved = p.move(chamber, jet)
            #print_chamber(chamber)
            #print('')
        h.append(max(chamber, key=lambda k: k[1])[1] + 1)
    i, di = get_pattern(h)
    dh = np.diff(h)
    height = calc_height(1000000000000, h[:i+1], dh[i:i+di])
    return height
```
