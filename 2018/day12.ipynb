```python
import numpy as np
```

```python
def read_input(filename):
    with open(filename, 'r') as f:
        state = f.readline().strip().split(' ')[2]
        state = np.array([1 if x == '#' else 0 for x in state])
        f.readline()
        grow = []
        ungrow = []
        for line in f.readlines():
            rule, res = line.strip().split(' => ')
            rule = [1 if x == '#' else 0 for x in rule]
            if res == '#':
                grow.append(rule)
            else:
                ungrow.append(rule)
        grow = np.array(grow)
        ungrow = np.array(ungrow)
    return state, grow, ungrow

def pad(x):
    pad_start = 3
    pad_end = 3
    for i in range(min(3, len(x))):
        if x[i] != 0:
            break
        pad_start -= 1
    for i in range(-1, -min(4, len(x) + 1), -1):
        if x[i] != 0:
            break
        pad_end -= 1
    pad_start = max(0, pad_start)
    pad_end = max(0, pad_end)
    return np.pad(x, pad_width=(pad_start, pad_end)), pad_start

def step(state, grow, ungrow, shift0):
    state, pad_shift = pad(state)
    shift0 = shift0 - pad_shift + 2
    new_state = np.zeros(state.size - 4)
    for i in range(len(new_state)):
        if np.any(np.all(state[i:i+5] == grow, axis=1)):
            new_state[i] = 1
        #elif np.any(np.all(state[i:i+5] == ungrow, axis=1)):
        #    new_state[i] = 0
    state = new_state
    return state, shift0

def print_state(state):
    s = ['#' if x == 1 else '.' for x in state]
    print(''.join(s))
```

```python
state, grow, ungrow = read_input('12_input.txt')
shift0 = 0
for i in range(20):
    state, shift0 = step(state, grow, ungrow, shift0)
sum([i + shift0 for (i, val) in enumerate(state) if val == 1])
```
