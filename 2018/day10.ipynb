```python
import re
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_input(filename):
    p = []
    v = []
    expr = re.compile(r'position=<(-|\W?\d*), (-|\W?\d*)> velocity=<(-|\W?\d*), (-|\W?\d*)>')
    with open(filename, 'r') as f:
        for line in f.readlines():
            m = expr.match(line.strip())
            p.append([int(m[1]), int(m[2])])
            v.append([int(m[3]), int(m[4])])
    return p, v
```

```python
def step(p, v):
    for pp, vv in zip(p, v):
        pp[0] += vv[0]
        pp[1] += vv[1]
    return p

def display(p):
    x_min = min(pp[0] for pp in p)
    x_max = max(pp[0] for pp in p)
    y_min = min(pp[1] for pp in p)
    y_max = max(pp[1] for pp in p)
    x = [pp[0] - x_min for pp in p]
    y = [pp[1] - y_min for pp in p]
    x_max -= x_min
    y_max -= y_min
    m = np.zeros((y_max + 1, x_max + 1))
    for xx, yy in zip(x, y):
        m[yy, xx] = 1
    plt.matshow(m)
```

```python
p, v = read_input('10_input.txt')
dx = max(pp[0] for pp in p) - min(pp[0] for pp in p)
dy = max(pp[1] for pp in p) - min(pp[1] for pp in p)
n = 0
while True:
    p_prev = p
    dx_prev = dx
    dy_prev = dy
    p = step(p, v)
    dx = max(pp[0] for pp in p) - min(pp[0] for pp in p)
    dy = max(pp[1] for pp in p) - min(pp[1] for pp in p)
    n += 1
    if dx > dx_prev or dy > dy_prev:
        print(f'n = {n}')
        break
```

```python
p, v = read_input('10_input.txt')
for _ in range(n - 5):
    p = step(p, v)
```
