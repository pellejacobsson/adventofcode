```python
import numpy as np
```

```python
def read_input(filename):
    p = []
    v = []
    a = []
    with open(filename, 'r') as f:
        for l in f.readlines():
            lp, lv, la = l.strip().split(', ')
            p.append([int(x) for x in lp[3:-1].split(',')])
            v.append([int(x) for x in lv[3:-1].split(',')])
            a.append([int(x) for x in la[3:-1].split(',')])
    return np.array(p), np.array(v), np.array(a)
            
```

```python
p, v, a = read_input('20_input.txt')
def closest(p, v, a, i):
    return np.argmin(np.sum(np.abs(p + i * v + i *(i + 1) / 2 * a), axis=1))
[closest(p, v, a, i) for i in np.logspace(1, 10, 10)]
```

```python
p, v, a = read_input('20_input.txt')
s = []
for n in range(1000):
    pp = p + n * v + n * (n + 1) / 2 * a
    ixs = []
    for i in range(p.shape[0]):
        if np.any(np.all(pp[i, :] == np.delete(pp, i, axis=0), axis=1)):
            ixs.append(i)
    p = np.delete(p, ixs, axis=0)
    v = np.delete(v, ixs, axis=0)
    a = np.delete(a, ixs, axis=0)
    s.append(p.shape[0])
```
