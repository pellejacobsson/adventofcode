```python
from math import floor
```

```python
def run_program(reg, program):
    i_halt = len(program)
    i = 0
    output = []
    while True:
        if i >= i_halt:
            break
        opcode = program[i]
        operand = program[i + 1]
        literal = operand
        match operand:
            case 0 | 1 | 2 | 3:
                combo = operand
            case 4:
                combo = reg[0]
            case 5:
                combo = reg[1]
            case 6:
                combo = reg[2]
        match opcode:
            case 0:
                reg[0] = floor(reg[0] / 2**combo)
                i += 2
            case 1:
                reg[1] = reg[1] ^ literal
                i += 2
            case 2:
                reg[1] = combo % 8
                i += 2
            case 3:
                if reg[0] != 0:
                    i = literal
                else:
                    i += 2
            case 4:
                reg[1] = reg[1] ^ reg[2]
                i += 2
            case 5:
                output.append(combo % 8)
                i += 2
            case 6:
                reg[1] = floor(reg[0] / 2**combo)
                i += 2
            case 7:
                reg[2] = floor(reg[0] / 2**combo)
                i += 2
    return output
```

```python
reg = [48744869, 0, 0]
program = [2, 4, 1, 2, 7, 5, 1, 3, 4, 4, 5, 5, 0, 3, 3, 0]
output = run_program(reg, program)
print(",".join([str(n) for n in output]))
```
